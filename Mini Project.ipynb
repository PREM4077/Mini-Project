```python
import keras_ocr
import matplotlib.pyplot as plt
pipeline=keras_ocr.pipeline.Pipeline()
```

```python
images=[keras_ocr.tools.read(img) for img in["C:\\Users\\premk\\OneDrive\\Desktop\\WhatsApp Image 2023-09-30 at 20.38.07.jpeg","C:\\Users\\premk\\OneDrive\\Desktop\\WhatsApp Image 2023-09-30 at 20.37.23.jpeg"]]
```

```python
prediction_groups=pipeline.recognize(images)
```

```python
ig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions,ax=ax)
```
